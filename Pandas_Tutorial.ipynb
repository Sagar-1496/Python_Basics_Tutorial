##Cleaning Data

```python
#Check for null values
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
#Fill null values
df.fillna(130, inplace=True)
df['Calories'].fillna(130, inplace = True)
```

```python
#fill null with mean()
x = df['Calories'].mean()
df['Calories'].fillna(x, inplace=True)
```
